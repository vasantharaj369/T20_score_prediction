```python
# importing the requirements

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from yaml import safe_load
from tqdm import tqdm
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
filenames = []
for file in os.listdir('data'):
    filenames.append(os.path.join('data',file))
```

```python
# creating a data frame from json files

final_df = pd.DataFrame()
counter = 1
for file in tqdm(filenames):
    with open(file, 'r') as f:
        df = pd.json_normalize(safe_load(f))
        df['match_id'] = counter
        final_df = pd.concat([final_df, df])
        counter+=1

final_df
```

```python
final_df.head()
```

```python
pickle.dump(final_df, open('converted_json.pkl', 'wb'))
```

```python
# dropping unwanted columns

final_df.drop(columns=[
    'meta.data_version',
    'meta.created',
    'meta.revision',
    'info.outcome.bowl_out',
    'info.bowl_out',
    'info.supersubs.South Africa',
    'info.supersubs.New Zealand',
    'info.outcome.eliminator',
    'info.outcome.result',
    'info.outcome.method',
    'info.neutral_venue',
    'info.match_type_number',
    'info.outcome.by.runs',
    'info.outcome.by.wickets'],inplace=True)
```

```python
df = final_df
df.head()
```

```python
df['info.gender'].value_counts()
```

```python
# removing the female matches

df = df[df['info.gender']=='male']
```

```python
df['info.overs'].value_counts()
```

```python
# removing the 50 over matches

df = df[df['info.overs']==20]
```

```python
df = df.reset_index(drop = True)
df.head()
```

```python
ball = []
bat = []
batting_team = []
bowler = []
run = []
player_dismissed = []
match_id = []
teams = []
citys = []
venues = []

count = 1
for i in range(len(df)):
    a = df.iloc[:, 0][i][0]['1st innings']['deliveries']
    batting =  df.iloc[:, 0][i][0]['1st innings']['team']
    team = df.loc[:, 'info.teams'][i]
    city = df.loc[:, 'info.city'][i]
    venue = df.loc[:, 'info.venue'][i]
    for j in range(len(a)):
        ball.append(list(a[j].keys())[0])
        bat.append(list(a[j].values())[0]['batsman'])
        bowler.append(list(a[j].values())[0]['bowler'])
        run.append(list(a[j].values())[0]['runs']['total'])
        match_id.append(count)
        teams.append(team)
        citys.append(city)
        
        batting_team.append(batting)
        venues.append(venue)
        try:
            player_dismissed.append(list(a[j].values())[0]['wicket']['player_out'])
        except:
            player_dismissed.append(0)
        
    count = count+1
```

```python
df1 = pd.DataFrame({
    'match_id': match_id,
    'teams': teams,
    'batting_team': batting_team,
    'ball': ball,
    'batsman': bat,
    'bowler': bowler,
    'runs': run,
    'player_dismissed': player_dismissed,
    'city': citys,
    'venue': venues
})

df1
```

```python
def bowling(row):
    for i in row['teams']:
        if i!= row['batting_team']:
            return i
df1['bowling_team'] = df1.apply(bowling, axis = 1)

df1 = df1.drop(columns = ['teams'])
df1.head()
```

```python
# removing the teams that have not played many matches

teams = [ 'Australia', 'India', 'Bangladesh', 'New Zealand', 'South Africa', 'England',
            'West Indies', 'Afghanistan', 'Pakistan', 'Sri Lanka']

df1 = df1[df1['batting_team'].isin(teams)]
df1 = df1[df1['bowling_team'].isin(teams)]
```

```python
df = df1[['match_id', 'batting_team', 'bowling_team', 'ball', 'runs', 'player_dismissed', 'city', 'venue']]
df.head()
```

```python
# checking for null values

df.isnull().sum()
```

```python
# replacing th null values in city with the venue

cities = np.where(df['city'].isnull(), df['venue'].str.split().str[0], df['city'])
df['city'] = cities
df.head()
```

```python
# removing the cities that has minimum number of matches played

cities = df['city'].value_counts()[df['city'].value_counts() > 600].index
df = df[df['city'].isin(cities)]
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
df.head()
```

```python
df = df.reset_index(drop = True)
df
```

```python
df['over'] = df['ball'].astype(str).str.split('.').str[0]
df['ball_no'] = df['ball'].astype(str).str.split('.').str[1]
df['balls_bowlled'] = ((df['over'].astype(int))*6) + df['ball_no'].astype(int)
df
```

```python
df['balls_remainning'] = 120 - df['balls_bowlled']
df['balls_remainning'] = df['balls_remainning'].apply(lambda x: 0 if x<0 else x)
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x: 0 if x == 0 else 1)
df.head()
```

```python
df['wickets_left'] = 10 - df.groupby('match_id')['player_dismissed'].cumsum()
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowlled']
df.head()
```

```python
df['run'] = 1
ids = df['match_id'].unique()

for i in ids:
    a = df.loc[df['match_id']==i].index[:30]
    a = a.values
    df.loc[a, 'run'] = 0
```

```python
for i in range(0, 51109, 1):
        if df['run'][i]==0:
            continue
        else:
            df['run'][i] = df['runs'][i-30 : i].sum()
            
df = df.rename(columns = {'run': 'last_five'})
df.head()
```

```python
df['last_five'] = df['last_five'].replace(0, np.nan)
df.head()
```

```python
df = df.groupby('match_id')['runs'].sum().reset_index().merge(df, on = 'match_id')
df.head()
```

```python
df = df[['batting_team', 'bowling_team', 'city', 'current_score', 'balls_remainning', 'wickets_left', 'crr', 'last_five', 'runs_x']]
```

```python
df.head()
```

```python
df = df.dropna().reset_index(drop = True)
df.head()
```

```python
X = df.drop(columns = ['runs_x'])
Y = df['runs_x']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge, Lasso
from sklearn.svm import SVR

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state= 0)
```

```python
X_train
```

```python
from sklearn.pipeline import Pipeline
```

```python
column = ColumnTransformer([('one', OneHotEncoder(sparse_output = False), ['batting_team', 'bowling_team', 'city'])], remainder= 'passthrough')
col = ColumnTransformer([('two', StandardScaler(), ['current_score', 'balls_remainning', 'wickets_left', 'crr', 'last_five'])], remainder= 'passthrough')

pipe = Pipeline(steps = [
    ('step1', column),
    ('step2', StandardScaler()),
    ('step3', LinearRegression())
])

pipe.fit(X_train, Y_train)
pre = pipe.predict(X_test)
print(r2_score(pre, Y_test))
print(mean_absolute_error(pre, Y_test))
```
